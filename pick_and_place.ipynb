```python
import gym
import gym_franka
env = gym.make('panda-v0')
for _ in range(50):
    observation = env.reset()
    for t in range(100):
        env.render()
        print(observation)
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        if done:
            print(f"Episode finished after {t + 1} timesteps")
            break
env.close()
```
